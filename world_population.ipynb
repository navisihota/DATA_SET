```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population_data=pd.read_csv('https://github.com/navisihota/Dataset/blob/main/world_population.csv?raw=True')
```

```python
population_data.head()
```

```python
population_data.info()
```

```python
population_data.isnull().sum()
```

```python
data=population_data.sort_values(by=['2022 Population'],ascending=False).head(10)
```

```python
data
```

```python
plt.figure(figsize=(16,5))
plt.plot(data['Country/Territory'],data['2022 Population'],label='Population',color='orange',marker='d',markersize=10,markeredgecolor='green')
plt.grid(alpha=0.5)
plt.xlabel('Country Name',size=18)
plt.ylabel('Population in 2022 in cr',size=18)
plt.legend()
plt.title('Top 10 Countrys Pololation in 2022',size=30,color='green')
```

```python
#first of all storing the data as an array in a variable from the previously created top_10 variable using iloc method
```

```python
a=data.iloc[1,5:12]
```

```python
plt.figure(figsize=(15,6))
plt.bar(a.index,a,width=0.5,color='#c9e52f' )
plt.title('Change in the population of India in multiple of cr',size=30,color='#7C9D96')
plt.xlabel('Year',size=24)
plt.ylabel('population',size=24)
plt.legend()
plt.show()
```

```python
# difference of population of china and india during the given years
```

```python
b=data.head(2)
b
```

```python
c=b.iloc[:2,5:12]
c
```

```python
#claculating the difference

diff=c.iloc[0,::]-c.iloc[1,::]
diff
```

```python
plt.figure(figsize=(12,6))
x=plt.axes()
x.set_facecolor('#F4D160')
plt.plot(diff.index,diff,color='blue',marker='d',markersize=8, markerfacecolor='lavender',markeredgecolor='black')
plt.bar(diff.index,diff,width=0.4,color='red',label='difference')
plt.title('Difference between the population of China and India ',size=24,color='lime')
plt.xlabel('years',size=24,color='#3A614B')
plt.ylabel('difference b/w population',size=24,color='#3A614B')
plt.grid(alpha=0.3)
plt.legend()
```
