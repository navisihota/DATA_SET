```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pokemon_data = pd.read_csv('https://github.com/navisihota/Dataset/blob/main/pokemon.csv?raw=True')
```

```python
pokemon_data.head()
```

```python
pokemon_data.info()
```

```python
pokemon_data['type1'].unique()
```

Attack of diffrent type of pokemons vs other

```python
plt.figure(figsize=(12,8))
sns.barplot(data=pokemon_data,x='type1',y='attack')
plt.show()
```

```python
pokemon_data['type1'].value_counts()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=pokemon_data,x='type1',palette='icefire')
```

```python
plt.figure(figsize=(12,8))
plt.pie(pokemon_data['type1'].value_counts(),labels=pokemon_data['type1'].value_counts().index)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=pokemon_data,x='against_ice',y='against_fire')
plt.show()
```

```python
pokemon_data.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(pokemon_data.corr())
```

```python
temp_data = pokemon_data.groupby(by='type1').mean()[['against_bug', 'against_dark', 'against_dragon',
       'against_electric', 'against_fairy', 'against_fight', 'against_fire',
       'against_flying', 'against_ghost', 'against_grass', 'against_ground',
       'against_ice', 'against_normal', 'against_poison', 'against_psychic',
       'against_rock', 'against_steel', 'against_water']]
```

```python
plt.figure(figsize=(17,8))
sns.heatmap(temp_data,annot=True)
```

Top 10 Attact and Defense Pokemons

```python
top_10=pokemon_data.sort_values(by=['attack','defense'],ascending=False).head(10)
top_10
```

```python
plt.figure(figsize=(12, 8))
x=plt.axes()
plt.bar(top_10['name'], top_10['attack'], color='lime',width=0.5, label='Attack')
plt.bar(top_10['name'],top_10['defense'], color='green',width=0.3, label='Defense')
plt.title('Top 10 Strongest PokÃ©mon (Attack and Defense)')
plt.legend()
```

```python
heavy=pokemon_data.sort_values(by=['weight_kg','is_legendary'],ascending=False).head(10)
heavy
```

```python
plt.figure(figsize=(12,8))
x=plt.axes()
plt.bar(heavy['name'],heavy['weight_kg'],color='green',width=0.6,label='weight')
plt.bar(heavy['name'],heavy['sp_attack'] ,color='red',width=0.3,label='sp_attack')
plt.ylabel('Names',size=15,color='purple')
plt.title('Heaviest Legendary Pokemon(weight and special attacks) stats',size=15,color='red')
plt.legend()
plt.show()
```
